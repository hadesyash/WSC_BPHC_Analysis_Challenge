# ***Getting daily returns:***

```python
def daily_returns():
  print("Daily return",df1["datetime"][0],"is","Null") #first value
  for j in range(1,3447):
    print("Daily return on",df1["datetime"][j],"is",df1["close"][j]-df1["close"][j-1])
daily_returns() 
```

# ***Getting outliers:***

```python
out_ind=[]
def get_outliers(df1):
  for i in range(1,3447):
    if ((df1["close"][i]-df1["close"][i-1])/(df1["close"][i-1])*100) > 5 or ((df1["close"][i]-df1["close"][i-1])/(df1["close"][i-1])*100) < -5:
      out_ind.append(i)
  print(out_ind)
get_outliers(df1)
```

# ***Removing outliers:***

```python
for k in out_ind:
  df1 = df1.drop(k)
```
