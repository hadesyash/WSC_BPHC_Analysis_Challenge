```python
#plot

years_list=[]
closing_price=[]
for i in range(0,3448):
  years_list.append(df1["datetime"][i])
  closing_price.append(df1["close"][i])
```

```python
import matplotlib.pyplot as plt

plt.plot(years_list,closing_price)

plt.xlabel("Dates")
plt.ylabel("Closing prices")
plt.title("Closing prices against datetime")

plt.show()
```
