```python
#daily percentage

for i in old_values:
  c=i['close']
  o=i['open']
  dp=round(((c-o)/o)*100,2)
  print("Daily percentage of",i['name'],"is",dp)
```

```python
#plotting highs

import numpy as np
import matplotlib.pyplot as plt

name_list=[]
high_list=[]
for k in old_values:
  name_list.append(k['name'])
  high_list.append(k['high'])

fig = plt.figure(figsize = (20, 10))
 
# creating the bar plot
plt.bar(name_list, high_list, color ='red',
        width = 0.4)
 
plt.xlabel("Names of stocks")
plt.ylabel("Highs of the stocks")
plt.title("Bar plot of highs of the stocks")
plt.show()
```
